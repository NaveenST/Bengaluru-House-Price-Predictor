```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df10 = pd.read_json("C:\\Users\\stnav\\Bengaluru Housing Price Estimator Project\\Code\\outliers_cleaned_dataset")
df10.head()
```

```python
dummies1 = pd.get_dummies(df10['location'], drop_first=True)
dummies1
```

```python
dummies2 = pd.get_dummies(df10['area_type'], drop_first=True)
dummies2
```

```python
dummies = dummies1.merge(dummies2, left_index=True, right_index=True)
dummies
```

```python
df11 = pd.concat([df10,dummies], axis='columns')
df11.head()
```

```python
df12 = df11.drop(['area_type','location'], axis='columns')
df12.head()
```

```python
df12.shape
```

```python
X = df12.drop('price', axis='columns')
X.head()
```

```python
y = df12['price']
y.head()
```

```python
X.shape
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
lg.fit(X_train,y_train)
lg.score(X_test,y_test)
```

### Base Model Score is 85.4% for Linear Regression

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=10)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

### Using 5 split cross validation we're able to get scores above almost 80% with 86% as the highest.

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

### Linear Regression Model is the best model

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lg.predict([x])[0]
```
